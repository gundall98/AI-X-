# 파이썬 numpy 모듈을 이용한 IRR 계산기 (연기준 현금흐름)
### 기업가치평가 강창모 교수님
## 필요 입력 데이터:
* 만기
* 0 시점의 투하자본 (양의 값)
* 연도 말마다의 현금 흐름
* 만일 매년 말의 현금 흐름이 같다면 진행에 따라 "1" 입력바람



```
import numpy as np
print("===================================")
print("Python IRR Calculator")
print("Author By: Park Gun")
print("HanYang Univ. Department Of Finance")
print("===================================")

#IRR 계산기
def IRRcal():
    # 만기 데이터 입력
    try:
      years = int(input("Please Input Maturity: \n"))
    except Exception:
      print("[Error] Maturity must be int not str")
      IRRcal()
    else:
      pass
    # 투하자본 데이터 입력
    try:
      IC = -int(input("\nPlease Input Ammount Of Initial Investment: \n-"))
    except Exception:
      print("[Error] Initial Investment must be int not str")
      IRRcal()
    else:
      pass

    # 현금흐름 저장
    CFs=[]
    # 현금흐름 입력
    for i in range(1,years+1):
      try:
        CF = int(input("\nPlease Input Cash Flows in years "+str(i)+" : \n"))
      except Exception:
        print("[Error] Cash Flows must be int not str")
        IRRcal()
      else:
        CFs.append(CF)
        if i >=1:
          if CF in CFs:
            check = input("\nIf the all left Cash Flows are same,\nPlease input '1', if not, please input'0'.\n")
            if check == "1":
              CFs = []
              for i in range(0,years):
                CFs.append(CF)
              break
            else:
              pass
        else:
          pass

    CFs.insert(0,IC)
    print()
    print("Cash Flows = "+str(CFs))
    print("IRR        = "+ str(round(np.irr(CFs),years)*100)+"%")
    print()
    continues = input("Press 'Enter' to continue.\n")
    if continues == "":
      IRRcal()
    else:
      pass
IRRcal()

```
